### 簡單線性回歸

```python
import numpy as np
import matplotlib.pyplot as plt
#直徑
X = np.array([[6],[8],[10],[14],[18]]).reshape(-1,1)
#重塑矩陣 reshape(-1,1) = 組成二維矩陣 ; -1 次特殊值自動計算矩陣長度
y = [7,9,13,17.5,18]
#價格
plt.figure()
# 生成畫布
plt.title("pizza price plotted against diameter")
plt.xlabel("Diameter in inches")
plt.ylabel("Price in dollars")
plt.plot(X,y,'k.')
plt.axis([0,25,0,25])
plt.grid(True)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
# LinearRegression需要 (數據,標籤) = (值, 截距)

test_pizza = np.array([[12]])#要二維矩陣符合模型要求(-1,1)
# 以比薩直徑12作為樣本
predicted_price = model.predict(test_pizza)[0]
#讓模型去計算對應的y值並提取index = 0的值
print('12" pizza should cost: $%.2f' % predicted_price)
```
